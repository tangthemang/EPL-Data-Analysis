```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://fbref.com/en/comps/9/shooting/Premier-League-Stats#stats_shooting'
```

```python
response = requests.get(url).text.replace('<!--', '').replace('-->', '')
df = pd.read_html(response, header=1)[2]
```

```python
df
```

```python
df1 = pd.DataFrame(columns=df.columns)
```

```python
rows_to_keep = []
for i in range(df.shape[0]):
    if str(df.iloc[i, 0]) != "Rk":
        rows_to_keep.append(i)

df1 = df.iloc[rows_to_keep, :].reset_index(drop=True)
df1['Gls'] = pd.to_numeric(df1['Gls'], errors='coerce')
df1 = df1.sort_values(['xG', 'Gls'])
df1['xG'] = df1['xG'].astype(float)
df1['Gls'] = df1['Gls'].astype(float)
plt = df1.plot.scatter(x='xG', y='Gls')

# Calculate the line of best fit using numpy's polyfit function
m, b = np.polyfit(df1['xG'], df1['Gls'], 1)

# Create a scatter plot of 'Dist' vs 'Gls'
ax = df1.plot.scatter(x='xG', y='Gls')

# Add the line of best fit to the plot using matplotlib's plot function
ax.plot(df1['xG'], m*df1['xG'] + b, color='red')
```

```python
slope = m

print("Slope of the line of best fit:", slope)
```

```python
df1['Sh'] = df1['Sh'].astype(int)
shots = df1['Sh'].sum()
print(shots)
```

```python
goals = df1['Gls'].sum()
exgoals = df1['xG'].sum()
print(goals)
print(exgoals)
print(goals/exgoals)
```

```python
df1['G/xG'] = np.nan

for i in range(df1.shape[0]):
    gls = df1.iloc[i, df1.columns.get_loc('Gls')]
    xg = df1.iloc[i, df1.columns.get_loc('xG')]
    if xg == 0:
        df1.iloc[i, df1.columns.get_loc('G/xG')] = np.nan
    else:
        if gls == 0:
            df1.iloc[i, df1.columns.get_loc('G/xG')] = 0
        else:
            df1.iloc[i, df1.columns.get_loc('G/xG')] = gls / xg
            print(f"Row {i}: gls = {gls}, xg = {xg}, G/xG = {gls/xg}")
```

```python
df2_rows = []
for i in range(df1.shape[0]):
    if df1.iloc[i, df1.columns.get_loc('Sh')] > 20:
        df2_rows.append(i)

df2 = df1.iloc[df2_rows, :].reset_index(drop=True)
```
